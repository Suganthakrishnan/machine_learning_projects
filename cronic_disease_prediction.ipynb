```python
cols_to_be_converted = ["packed cell vol","wbc count","rbc count"]

def object_to_numeric(df,cols_to_be_converted):
    for columns in cols_to_be_converted:
        df[columns] = pd.to_numeric(df[columns],errors="coerce")
        print(f"the converted column is {columns} and the datatype is{df[columns].dtype}")

object_to_numeric(df,cols_to_be_converted)
```

```python
new_cols=[]
for i in df.columns:
    new_cols.append(i)

print(new_cols)
```

```python
#handling missing values

missing = df.isnull().sum().sort_values(ascending=False)
missing
# now for numeric columns we are going to fill mean and for object ccolumns we are going to fill with mode

def fill_mean(df,column):
    mean_value = df[column].mean()
    df[column].fillna(mean_value,inplace = True)

def fill_mode(df,column):
    mode_value = df[column].mode()[0]
    df[column].fillna(mode_value,inplace = True)

num_col = []


for i in new_cols:
    if df[i].dtype != object:
        num_col.append(i)
    
# using list comprehension also we can do the above step
obj_col = [i for i in new_cols if df[i].dtype == 'object']
obj_col


print(num_col)
print(obj_col)
for i in num_col:
    fill_mean(df,i)

for i in obj_col:
    fill_mode(df,i)

df.head(20)

```

```python
missing = df.isnull().sum().sort_values(ascending=False)
missing
```

```python
# gotta check the no of categories in each object columns
for i in obj_col:
    print(f"in the column {i}, these are the categories: {df[i].unique()}")
```

```python
# gotta manage the categories in the following columns : diabetes mellitus , artery diseases, classification
df["diabetes mellitus"] = df["diabetes mellitus"].replace(to_replace={" yes":"yes",'\tno':"no","\tyes":"yes"})
df["coronary artery disease"] = df["coronary artery disease"].replace(to_replace = {"\tno":"no"})
df["classification"] = df["classification"].replace({"ckd\t":"ckd"})

for i in obj_col:
    print(f"after correction the categories present column {i} are {df[i].unique()}")
```

```python
from category_encoders import BinaryEncoder

my_encoder = BinaryEncoder(cols=obj_col)

df = my_encoder.fit_transform(df)

df

cols_to_be_dropped = [cols for cols in df if cols.endswith('_1')]
cols_to_be_dropped

df = df.drop(columns=cols_to_be_dropped)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(13,9))
sns.heatmap(df.corr(),annot = True)
plt.show()
```

```python
correlation = df.corr()["classification_0"].abs().sort_values(ascending = False)[1:]  # we use .abs() to sort the values and find the most correlated value for our featre selection
print(correlation)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(columns = "classification_0")
y = df["classification_0"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

dct = DecisionTreeClassifier()

model = dct.fit(x_train,y_train)



y_predict = model.predict(x_test)
y_predict
```

```python
# testing the model via metrices
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,mean_squared_error

print("the analysis of the model are as follows: \n")
print("accuracy:",accuracy_score(y_predict,y_test),'\n')
print("precision:",precision_score(y_predict,y_test),'\n')
print("recall:",recall_score(y_predict,y_test),'\n')
print("f1_score:",f1_score(y_predict,y_test),'\n')

```
